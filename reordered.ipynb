```python
order_products = pd.read_csv('data/order_products.csv')
```

```python
orders = pd.read_csv('data/orders.csv')
```

```python
data = pd.merge(order_products, orders, on = "order_id",  how = "left")
```

```python
data.shape
```

```python
grouped = data.groupby(['user_id', 'product_id'])["order_id"].aggregate("min").reset_index()
grouped
```

```python
grouped['indicator']=0
```

```python
grouped.shape
```

```python
result = pd.concat([order_products, grouped], axis=1, join='outer')
result.head()
```
